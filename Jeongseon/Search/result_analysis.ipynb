```python
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
# CSV 파일 로드
df_original = pd.read_csv("melb_split.csv")
df_all = pd.read_csv("optimized_all.csv")  # 전체 최적화 결과
df_row = pd.read_csv("optimized_solutions.csv")  # 행별 최적화 결과
```

```python

# 📌 1️⃣ 최적화된 위치 시각화 (위도-경도 산점도)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_original["Longtitude"], df_original["Lattitude"], c=df_original["Price"], cmap='viridis', alpha=0.8)
plt.colorbar(scatter, label="Price")
plt.xlabel("Longtitude (경도)")
plt.ylabel("Lattitude (위도)")
plt.title("📍 최적화된 위치와 예측 가격 (Scatter Plot)")
plt.show()
```

```python

# 📌 1️⃣ 최적화된 위치 시각화 (위도-경도 산점도)
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_row["Longtitude"], df_row["Lattitude"], c=df_row["Predicted_Price"], cmap='viridis', alpha=0.8)
plt.colorbar(scatter, label="Predicted Price")
plt.xlabel("Longtitude (경도)")
plt.ylabel("Lattitude (위도)")
plt.title("📍 최적화된 위치와 예측 가격 (Scatter Plot)")
plt.show()
```

```python
# 📌 2️⃣ 예측 가격 분포 확인 (히스토그램)
plt.figure(figsize=(10, 6))
sns.histplot(df_original["Price"], bins=30, kde=True, color='blue')
plt.xlabel("Predicted Price (예측된 가격)")
plt.ylabel("Frequency (빈도)")
plt.title("📊 예측된 가격의 분포 (Histogram)")
plt.show()
```

```python
# 📌 2️⃣ 예측 가격 분포 확인 (히스토그램)
plt.figure(figsize=(10, 6))
sns.histplot(df_row["Predicted_Price"], bins=30, kde=True, color='blue')
plt.xlabel("Predicted Price (예측된 가격)")
plt.ylabel("Frequency (빈도)")
plt.title("📊 예측된 가격의 분포 (Histogram)")
plt.show()
```

```python
# 📌 3️⃣ 3D 산점도 (위도, 경도, 예측된 가격)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df_original["Longtitude"], df_original["Lattitude"], df_original["Price"], c=df_original["Price"], cmap='plasma', alpha=0.8)
ax.set_xlabel("Longtitude (경도)")
ax.set_ylabel("Lattitude (위도)")
ax.set_zlabel("Price (가격)")
plt.colorbar(scatter, label="Price")
plt.title("🌍 공간적 가격 변화 (3D Scatter Plot)")
plt.show()
```

```python
# 📌 3️⃣ 3D 산점도 (위도, 경도, 예측된 가격)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df_row["Longtitude"], df_row["Lattitude"], df_row["Predicted_Price"], c=df_row["Predicted_Price"], cmap='plasma', alpha=0.8)
ax.set_xlabel("Longtitude (경도)")
ax.set_ylabel("Lattitude (위도)")
ax.set_zlabel("Predicted Price (예측된 가격)")
plt.colorbar(scatter, label="Predicted Price")
plt.title("🌍 공간적 가격 변화 (3D Scatter Plot)")
plt.show()
```

```python

# 📌 4️⃣ 위도별 예측 가격 변화 확인 (선 그래프)
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_original["Lattitude"], y=df_original["Price"], marker='o', color='red')
plt.xlabel("Lattitude (위도)")
plt.ylabel("Predicted Price (예측된 가격)")
plt.title("📈 위도별 예측 가격 변화 (Line Plot)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 📌 기존 데이터(melb_split.csv) 로드 (인덱스를 유지)
melb_df = pd.read_csv("melb_split.csv")

# 📌 최적화된 가격 데이터(optimized_solutions.csv) 로드
opt_df = pd.read_csv("optimized_solutions.csv")

# 🔹 기존 데이터에 인덱스 추가 (0부터 시작)
melb_df = melb_df.reset_index()

# 🔹 최적화 데이터와 기존 데이터(인덱스 기준) 병합
merged_df = opt_df.merge(melb_df, left_on="index", right_on="index", how="left")

# 🔹 가격 변화 계산
merged_df["Price_Difference"] = merged_df["Predicted_Price"] - merged_df["Price"]
merged_df["Price_Percentage_Change"] = (merged_df["Price_Difference"] / merged_df["Price"]) * 100

####################추가##################
# 🔺🔻 가격 변화 아이콘 추가
def add_arrow(value):
    if value > 0:
        return f"🔺 +{value:,.0f}"
    elif value < 0:
        return f"🔻 {value:,.0f}"
    else:
        return "→ 0"

merged_df["Change_Indicator"] = merged_df["Price_Difference"].apply(add_arrow)
###############################################################################

# ✅ 데이터 확인
print(merged_df[["index", "Lattitude_optimized", "Longtitude_optimized", "Lattitude", "Longtitude", "Price", "Predicted_Price", "Price_Difference", "Price_Percentage_Change"]])
```

```python
# 📊 1️⃣ 가격 변화 분포 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(merged_df["Price_Difference"], bins=30, kde=True, color='blue')
plt.axvline(0, color='red', linestyle='dashed')  # 가격 변화 0인 기준선
plt.xlabel("Price Difference (최적화된 가격 - 기존 가격)")
plt.ylabel("Frequency")
plt.title("📊 가격 변화 분포 (상승 vs 하락)")
plt.show()
```

```python
# 📊 2️⃣ 위도별 가격 변화 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x=merged_df["Lattitude"], y=merged_df["Price_Difference"], hue=merged_df["Price_Difference"], palette="coolwarm", size=abs(merged_df["Price_Difference"]))
plt.axhline(0, color='black', linestyle='dashed')  # 기준선
plt.xlabel("Lattitude (위도)")
plt.ylabel("Price Difference")
plt.title("📍 위도별 가격 변화")
plt.show()
```

```python
# 📊 3️⃣ 경도별 가격 변화 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x=merged_df["Longtitude"], y=merged_df["Price_Difference"], hue=merged_df["Price_Difference"], palette="coolwarm", size=abs(merged_df["Price_Difference"]))
plt.axhline(0, color='black', linestyle='dashed')  # 기준선
plt.xlabel("Longtitude (경도)")
plt.ylabel("Price Difference")
plt.title("📍 경도별 가격 변화")
plt.show()
```
