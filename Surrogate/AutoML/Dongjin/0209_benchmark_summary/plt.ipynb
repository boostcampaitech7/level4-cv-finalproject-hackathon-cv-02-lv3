```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

matplotlib.rcParams.update({'font.size': 14})
```

```python
def plot(dataset, metric, placement, models, results):
    # 예제 데이터 생성
    df = pd.DataFrame({"models": models, "results": results})

    # Seaborn 스타일 설정 (배경, 폰트 크기 등)
    sns.set(style="whitegrid", context="talk")

    # Figure 크기 설정
    plt.figure(figsize=(6, 4))

    # barplot 그리기: palette 옵션으로 색상을 다양하게 지정할 수 있음
    ax = sns.barplot(x='models', y='results', data=df, palette="hls", width=0.5)

    # 제목 및 축 레이블 추가
    plt.title(f"{dataset} dataset", fontsize=20)
    plt.xlabel("")
    plt.ylabel(metric)
    ax.yaxis.set_major_locator(ticker.MaxNLocator(3))

    if metric != "RMSE":
        plt.ylim([0, 1.1])
        plt.yticks([0, 0.5, 1])
    else:
        plt.ylim([0, 20])

    for p in ax.patches:
        # 막대의 높이 (y값)
        height = p.get_height()
        # 막대의 중앙 위치 (x값)
        x = p.get_x() + p.get_width() / 2
        # 막대 위에 텍스트 추가 (여기서는 소수점 없이 정수형태로 표시)
        ax.text(x, height, f'{height:.2f}', ha='center', va='bottom')


    # 그래프 출력
    sns.set(style="whitegrid", context="talk")
    ax.yaxis.grid(False)
    plt.show()

```

```python
df = pd.read_csv("benchmark.csv")
models = ['Ours', 'Auto-sklearn', 'TPOT']

for i, row in df.iterrows():
    dataset = row['Dataset']
    metric = row['Metric']
    placement = row['Placement (ours)']
    results = [row[model] for model in models]

    plot(dataset, metric, placement, models, results)
    
    

```
