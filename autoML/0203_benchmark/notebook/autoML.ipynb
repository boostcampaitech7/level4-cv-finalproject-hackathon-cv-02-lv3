```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
path = '../result/AutoML/parallel.csv'
df = pd.read_csv(path)
select_loc = range(3, 13)
```

```python
n_jobs = df['Unnamed: 0'].loc[select_loc].values.astype(float)
test_score = df['test_score'].loc[select_loc].values.astype(float)
test_score_std = df['test_score.1'].loc[select_loc].values.astype(float)
elapsed_time = df['elapsed_time'].loc[select_loc].values.astype(float)
elapsed_time_std = df['elapsed_time.1'].loc[select_loc].values.astype(float)
```

```python
matplotlib.rcParams.update({'font.size': 12})

plt.errorbar(n_jobs, elapsed_time, yerr=elapsed_time_std, fmt='o',                   
             capsize=5,                       
             label='Data with errorbars')  

plt.xlabel("Number of cpu cores")
plt.ylabel("Elapsed time (s)")
plt.title("CPU Cores vs Elapsed Time")
plt.ylim([0, 750])
plt.yticks([0, 200, 400, 600, 800])
```
